# Regular Expressions

```python
* pattern-format which repeats
* pattern[re] represents the set of values that match the pattern
* ^[0-9]$ -this pattern is for single digit numbers
* ^[0-9]{5}$-this  pattern is for selecting five digit numbers
* ^[a-z]{5}$-this pattern is for selecting five digit alphabets
* ^[a-z]$-this pattern is for selecting single alphabets
* ^[a-z0-9]{5}$-this pattern is for selecting both alphabets and numbers
* ^[6-9][0-9]{9}$-for mobile number validation
* ^[6-9][0-9]{9}|[+][9][1][6-9][0-9]{6}$-phone number validation
* ^[a-z][a-z0-9-.]{8,20}[@][a-z]{4,7}[.][a-z]{2,4}$-for gmail validation.
* for regular expreassion checking we can use "regex".
```

# Map functions

* syntax: map(iterable,function)


```python
def sum_N(n):
    s=0
    while(n>0):
            r=n%10
            s=s+r
            n=n//10
    print("sum is:",s)
sum_N(n)
li = [121,4343,65,77]
a=map(sum_N,li)
list(a)
```

```python
def evenorodd(n):
    if n%2==0:
        print("given",n,"is even number")
    else:
        print("given",n," is odd number")
li=[220,22,39]
b=filter(evenorodd,li)
list(b)
```

```python
def isnum(n):
    n=str(n)
    if n.isdigit():
        return True
    return False
li=["10","20","25","a"]
a=map(isnum,li)
list(a)
```

```python
a=filter(isnum,li)
list(a)
```

### File handling

* by using file handling we can write,read,append in python 

* performing operations on a file is called file handling
* entering date to a file is called as write to a file
* we can append,delete operations
* we can using "w" for write and "r"for retrive file

```python
def write_afile(filename,data):
    with open(filename,"w") as f:
        f.write(data)
    return
write_afile("sample.txt","a=int(input())\nb=1\nfor i in range(1,a+1):\nb *=i\nprint(b)\n")
```

```python
def readfromafile(filename):
    with open(filename,"r") as f:
        fd = f.read()
        f.close()
    return fd
readfromafile("sample.txt")
def printlines(filename):
     with open(filename,"r") as f:
        for line in f:
            print(line,end=" ")
        return
filename = "sample.txt"
#printlines(filename)
print(readfromafile(filename))
    
```

```python
### Append to a file:
def append_to_afile(filename,data):
    with open(filename,"a") as f:
        for line in data:
            f.write("\n"+line)
        return
data = ["sai","ram","kosana"]
filename = "sample.txt"
append_to_afile(filename,data)
```
